## Understanding `.loc` with a MultiIndex DataFrame

This DataFrame was created using `pd.concat(data)`, which results in a **MultiIndex**:

- **Level 0 (outer index):** data type (e.g., `'earnings'`, `'ret_prev'`, `'pe'`, `'analyst_est'`)
- **Level 1 (inner index):** dates (e.g., `'20210107'`)

Each row is identified by a tuple:
`('ret_prev', '20210107')`

---

### Command Breakdown (following cell)

`multi.loc[('ret_prev','20210107'), ['SPY','TLT']]`

#### 1. `.loc`
- Label-based indexing (uses index names, not positions)

#### 2. Row Selection
`('ret_prev','20210107')`

- Selects the row where:
  - Level 0 = `'ret_prev'`
  - Level 1 = `'20210107'`

#### 3. Column Selection
`['SPY','TLT']`

- Selects only these two columns from that row

---

### Result

Returns the values for SPY and TLT for `'ret_prev'` on `'20210107'`:

```
SPY    1.262910
TLT    0.941813
```

---

### Key Takeaways

- `.loc` uses **labels**, not positions
- MultiIndex rows require a **tuple**: `(level_0, level_1)`
- Columns are selected like a normal list

````python
# multi.loc[('ret_prev','20210107'), ['SPY','TLT']]
#
# Explanation:
# This DataFrame uses a MultiIndex created by pd.concat(data), where:
#   - Level 0 index = data type (e.g., 'earnings', 'ret_prev', etc.)
#   - Level 1 index = date (e.g., '20210107')
#
# ('ret_prev','20210107') is a tuple that selects a specific row:
#   - 'ret_prev' → first index level
#   - '20210107' → second index level
#
# ['SPY','TLT'] selects only those columns from that row
#
# Result:
# Returns the SPY and TLT values for 'ret_prev' on '20210107'
#
# Key idea:
#   - .loc = label-based indexing
#   - MultiIndex requires tuple-based row selection

multi.loc[('ret_prev','20210107'),['SPY','TLT']]
````

### Understanding `ser.unstack(level=0)`

```python
ser.unstack(level=0)
```

---

### What Was Passed

- `ser` is a **Pandas Series**
- It has a **MultiIndex** with two levels:
  - **Level 0:** category (e.g., `'earnings'`, `'ret_prev'`, `'pe'`, `'analyst_est'`)
  - **Level 1:** date (e.g., `'20210107'`)

---

### What `unstack(level=0)` Does

- Moves **Level 0 of the index** into **columns**
- Converts the Series into a **DataFrame**

#### Concept:
```
(category, date) → value
```

becomes:

```
date → columns (category)
```

---

### Before vs After

#### Before (Series)
```
earnings    20210105    0.87
ret_prev    20210105   -0.64
pe          20210105   -0.08
...
```

#### After (`unstack(level=0)`)
```
            earnings   ret_prev   pe   analyst_est
20210105     0.87      -0.64    -0.08    ...
20210106     ...
```

- **Index:** dates
- **Columns:** categories

---

### Why This Is Useful

- Converts data into a **tabular format**
- Ideal for:
  - Time series analysis
  - Feature engineering
  - Plotting
  - Machine learning inputs

---

### Other Options You Could Use

#### 1. Unstack a Different Level
```python
ser.unstack(level=1)
```
- Moves **dates into columns**
- Rows become categories

---

#### 2. Use Level Name Instead of Number
```python
ser.unstack(level='category')
```
- Same as `level=0`, but clearer

---

#### 3. Default Behavior
```python
ser.unstack()
```
- Defaults to the **last level** (`level=-1`)
- Equivalent to `level=1` in this case

---

#### 4. Fill Missing Values
```python
ser.unstack(level=0, fill_value=0)
```
- Replaces missing combinations with a value instead of NaN

---

#### 5. Multiple Levels (Advanced)
```python
ser.unstack(level=[0, 1])
```
- Used when there are more than two index levels

---

### Important Constraint

- Each index combination must map to **one value**
- If duplicates exist, Pandas will raise an error:
```
ValueError: Index contains duplicate entries
```

---

### Key Takeaways

- `unstack` moves an index level into columns
- `level=0` → first index level becomes columns
- Converts a **Series → DataFrame**
- Useful for reshaping data into analysis-ready format

````python
# unstack a multindex series into a dataframe
ser.unstack(level=0)
````

````python
ser_level1 = ser.unstack(level=1)
ser_level1
````

### Understanding `ser.unstack(level=0).stack()`

```python
ser.unstack(level=0).stack()
```

---

### What This Does

This is a **reshape → reshape back** operation:

1. `unstack(level=0)` converts part of the index into columns
2. `stack()` converts those columns back into the index

---

### Step 1: `unstack(level=0)`

- Moves **Level 0 (category)** into **columns**
- Converts the Series into a DataFrame

#### Before (Series)
```
(category, date) → value
```

#### After (DataFrame)
```
date → columns (category)
```

---

### Step 2: `.stack()`

- Moves **columns back into the index**
- Converts the DataFrame back into a Series

#### Result
```
(category, date) → value
```

---

### Final Outcome

- Returns a Series that **looks like the original**
- Often identical, but not always

---

### Important Caveats

#### 1. Missing Values (NaN)

- `unstack()` introduces `NaN` for missing combinations
- `stack()` **drops NaN values by default**

This means data can be lost during the round trip

---

#### 2. Index Ordering

- The order of index levels or values may change

---

### Safe Round Trip

To preserve all data:

```python
ser.unstack(level=0).stack(dropna=False)
```

---

### Why This Is Useful

#### 1. Reshaping for Analysis

```python
df = ser.unstack(level=0)
# perform analysis or transformations
ser_clean = df.stack()
```

---

#### 2. Data Cleaning

- `unstack()` exposes missing data
- `stack()` can remove incomplete rows

---

#### 3. Structure Normalization

- Helps enforce consistent MultiIndex structure

---

### Mental Model

- `unstack` → index → columns
- `stack` → columns → index

So this operation is:

```
index → columns → index
```

---

### Key Takeaways

- This is a **round-trip reshape**
- It often returns the original Series
- It may:
  - drop missing values
  - change ordering
- Useful for **reshaping and cleaning data**

````python
ser.unstack(level=0).stack()
````

### Understanding `pd.concat(data, axis=1)`

```python
multi = pd.concat(data, axis=1)
```

---

### What This Does

- Combines multiple DataFrames stored in `data`
- Concatenates them **horizontally (side-by-side)**

---

### Key Parameter

- `axis=1` → concatenate along **columns**

| Axis | Meaning |
|------|--------|
| `axis=0` (default) | Stack vertically (add rows) |
| `axis=1` | Stack horizontally (add columns) |

---

### Input

- `data` is typically a **dictionary of DataFrames**

Example:
```python
{
    'earnings': df1,
    'ret_prev': df2,
    'pe': df3
}
```

---

### Output

- A single **DataFrame**
- Columns become a **MultiIndex**

Example structure:

```
        earnings        ret_prev        pe
           SPY   TLT      SPY   TLT   SPY   TLT
date
20210105   ...   ...      ...   ...   ...   ...
```

---

### When to Use

- Combine multiple datasets with the **same index**
- Build feature tables (common in data science / ML)
- Align data from different sources side-by-side

---

### Key Takeaways

- `pd.concat(..., axis=1)` = horizontal merge
- Produces a **MultiIndex on columns**
- Requires keyword arguments in modern Pandas

````python
multi = pd.concat(data,axis=1)
````

# Pandas Indexing: Single vs Double Brackets (MultiIndex Case)

## 🧠 The Confusion

Example:

    print(type(multi['pe']))
    print(type(multi[['pe']]))

Output:

    <class 'pandas.DataFrame'>
    <class 'pandas.DataFrame'>

Normally:
- df['col'] → Series
- df[['col']] → DataFrame

But here, both return a DataFrame.

---

## 🔥 Root Cause: MultiIndex Columns

Your DataFrame has hierarchical (MultiIndex) columns.

Conceptually, instead of:

    ['pe', 'earnings']

You actually have:

    ('pe', 'SPY')
    ('pe', 'TLT')
    ('earnings', 'SPY')
    ('earnings', 'TLT')

So columns are grouped by levels.

---

## 🧠 What Pandas Is Actually Doing

### Case 1:

    multi['pe']

Interpretation:
- Select all columns where level 0 == 'pe'

Result:
- Multiple columns returned
- Therefore → DataFrame

---

### Case 2:

    multi[['pe']]

Interpretation:
- Select a list of top-level groups

Result:
- Still multiple columns
- Therefore → DataFrame

---

## ⚠️ Why the Expectation Breaks

Assumption:

    multi['pe'] → one column → Series

Reality:

    multi['pe'] → multiple columns → DataFrame

Because:
- 'pe' is NOT a single column
- It is a label in a column level

---

## 🧠 Updated Mental Model

Normal DataFrame:

    columns = ['pe', 'earnings']

→ 'pe' returns a Series

MultiIndex DataFrame:

    columns = [
        ('pe', 'SPY'),
        ('pe', 'TLT'),
        ...
    ]

→ 'pe' returns multiple columns → DataFrame

---

## 🛠️ How to Get a Series

You must fully specify the column:

    multi[('pe', 'SPY')]

or:

    multi['pe']['SPY']

Now:
- One column selected
- → Series

---

## ⚡ Alternative (Explicit Level Selection)

    multi.xs('pe', axis=1, level=0)

- Selects all 'pe' columns
- Returns DataFrame (still multiple columns)

---

## 🧭 Final Rule

Do NOT rely on bracket syntax alone.

Instead:

- 1 column selected → Series
- Multiple columns selected → DataFrame

---

## 💡 Key Insight

With MultiIndex:
- Labels like 'pe' refer to groups, not individual columns
- Output type depends on how many columns are returned

# Pandas MultiIndex: `swaplevel` vs `swaplevel().sort_index()`

## 🧠 Overview

Two similar-looking operations:

    multi.swaplevel(0, 1, axis=1)

and

    multi.swaplevel(0, 1, axis=1).sort_index(axis=1)

They produce **different usability outcomes**, even though both run successfully.

---

## 🔄 What `swaplevel` Does

    multi.swaplevel(0, 1, axis=1)

- Swaps levels in a MultiIndex (here, column levels 0 and 1)
- Changes the **hierarchical structure**, but **NOT the order**

### Example (Before)

    ('pe', 'SPY')
    ('pe', 'TLT')
    ('earnings', 'SPY')
    ('earnings', 'TLT')

### After `swaplevel`

    ('SPY', 'pe')
    ('TLT', 'pe')
    ('SPY', 'earnings')
    ('TLT', 'earnings')

---

## ⚠️ Problem After `swaplevel`

- Columns are now **out of order**
- Not grouped logically
- Harder to read and work with

Example:

    ('SPY', 'pe')
    ('TLT', 'pe')
    ('SPY', 'earnings')
    ('TLT', 'earnings')

This is valid, but messy.

---

## 🔽 What `sort_index` Fixes

    multi.swaplevel(0, 1, axis=1).sort_index(axis=1)

- Sorts columns after swapping
- Restores logical grouping

### Result:

    ('SPY', 'earnings')
    ('SPY', 'pe')
    ('TLT', 'earnings')
    ('TLT', 'pe')

---

## 🧪 Comparison

### Without Sorting

    multi.swaplevel(0, 1, axis=1)

- Structure changed ✔
- Order messy ❌

---

### With Sorting

    multi.swaplevel(0, 1, axis=1).sort_index(axis=1)

- Structure changed ✔
- Order clean ✔

---

## 🧠 Why Pandas Doesn’t Auto-Sort

Pandas assumes you may want to preserve original ordering.

So it:
- Changes structure
- Leaves ordering alone

---

## ⚡ Practical Impact

Without sorting:
- Column grouping is inconsistent
- Selection becomes confusing
- Harder to reason about structure

With sorting:
- Columns are grouped logically
- Easier to select and analyze data

---

## 🧭 Key Takeaway

- `swaplevel` → changes hierarchy
- `sort_index(axis=1)` → restores order

---

## 💡 Mental Model

Think of it like reorganizing a toolbox:

- `swaplevel` → dump tools into new drawers
- `sort_index` → actually organize them so you can find things again

````python
# rearrange axes
multi.swaplevel(0,1,1)
````

````python
multi.swaplevel(0,1,1).sort_index(axis=1)
````

# Pandas MultiIndex Aggregation Across Columns

## 🧠 Goal

Sum values across columns that share the same level in a MultiIndex.

---

## 🔧 Command

    multi.T.groupby(level=0).sum().T

---

## 🧠 What It Does

This operation groups columns by a specified level and sums values across those groups.

---

## 🔍 Step-by-Step Breakdown

### 1. Transpose the DataFrame

    multi.T

- Converts columns into rows
- MultiIndex columns become a MultiIndex index

---

### 2. Group by Level

    .groupby(level=0)

- Groups data based on level 0 of the index (original column level)
- Each group represents columns that shared the same top-level label

---

### 3. Aggregate

    .sum()

- Sums values within each group
- Reduces multiple entries into a single aggregated value per group

---

### 4. Transpose Back

    .T

- Restores original orientation
- Grouped results are now back in column form

---

## 🧪 Example Concept

Original column structure:

    ('pe', 'SPY')
    ('pe', 'TLT')
    ('earnings', 'SPY')
    ('earnings', 'TLT')

After operation:

    pe         → sum of all 'pe' columns
    earnings   → sum of all 'earnings' columns

---

## 🧭 Result

- Columns are grouped by level 0
- Values are aggregated across each group
- Output is a cleaner, reduced DataFrame

---

## 💡 Key Idea

To aggregate across MultiIndex columns:
- Temporarily treat columns as rows
- Perform grouping and aggregation
- Restore original structure

````python
# sum across an axis
multi.T.groupby(level=0).sum().T
````
