# Groupby, Pivot, Merge

```python
import pandas as pd 
import numpy as np 

univ = ['AAPL','MSFT','BAC','GS']
dates = pd.date_range('20110101','20201231')
df = pd.DataFrame(np.random.randn(len(dates),len(univ)),index=dates,columns=univ)
df
```

```python
sector = {'AAPL':'Tech','MSFT':'Tech','BAC':'Fin','GS':'Fin'}
data=[]
for dt in dates:
    for x in univ:
        data.append([x,sector[x],dt,np.random.randn(),np.random.randn()])
data = pd.DataFrame(data,columns=['ticker','sector','date','signal1','signal2'])
data
```

## Groupby

```python
# whats the average signal of each ticker?
data.groupby(['ticker', 'signal1', 'signal2']).mean(numeric_only=True)
```

```python
# whats the average signal of each sector on each day?
data.groupby(['sector','date']).mean(numeric_only=True)
```

```python
# can index before applying the function
data.groupby(['sector','date'])[['signal1']].mean(numeric_only=True)
```

# Pandas `.apply()` with `groupby`: Conceptual Summary

## 🧠 Core Idea

`.apply()` lets you run a **custom function** on each group in a `groupby` operation.

Unlike built-in methods (`mean`, `sum`, etc.), it gives you full control over the logic.

---

## 🔧 Example

    def max_minus_min(x):
        return x.max() - x.min()

    data.groupby(['sector','date'])[['signal1','signal2']].apply(max_minus_min)

---

## 🧠 What’s Happening

1. Data is split into groups based on:

       ('sector', 'date')

2. For each group:

       x = subset of the DataFrame

3. Your function runs on each group:

       x.max() - x.min()

4. Results are combined into a new DataFrame

---

## ⚡ When to Use `.apply()`

Use `.apply()` when:

- You need **custom logic** not available in built-in functions
- You want to operate on **entire groups**, not just columns
- You need flexibility beyond `.agg()` or `.transform()`

---

## ⚠️ Important Considerations

- Slower than built-in aggregations (`mean`, `sum`, etc.)
- Should not be your first choice if a built-in method exists
- Best used for **non-standard or complex operations**

---

## 🧭 Mental Model

Think of `groupby().apply()` as:

> "Split → Apply custom function → Combine results"

---

## 🔑 Key Insight

- `.agg()` → predefined, optimized operations
- `.transform()` → returns same shape as input
- `.apply()` → **maximum flexibility**, less performance

---

## 💡 Final Takeaway

`.apply()` is the “escape hatch” for group operations:

> Use it when Pandas doesn’t already give you the function you need

```python
# use apply with groupby to pass an arbitrary function
def max_minus_min(x):
    return x.max()-x.min()

data.groupby(['sector','date'])[['signal1','signal2']].apply(max_minus_min
```

# Pandas `pivot_table`: Conceptual Summary

## 🧠 Core Idea

`pivot_table` reshapes data from **long format → wide format**, while optionally aggregating duplicate values.

---

## 🔧 Example

    data.pivot_table(
        index='date',
        columns='sector',
        values='signal1',
        aggfunc='mean'
    )

---

## 🧠 What It Does

- Rows (`index`) → unique dates
- Columns (`columns`) → unique sectors
- Values (`values`) → aggregated `signal1` values
- Aggregation (`aggfunc`) → how duplicates are combined

---

## 🔄 Before (Long Format)

| date       | sector | signal1 |
|------------|--------|---------|
| 2021-01-01 | Tech   | 0.5     |
| 2021-01-01 | Tech   | 0.7     |
| 2021-01-01 | Energy | 0.2     |

---

## 🔁 After (Wide Format)

| date       | Tech | Energy |
|------------|------|--------|
| 2021-01-01 | 0.6  | 0.2    |

---

## ⚡ When to Use

Use `pivot_table` when:

- You need to **reshape raw data into a matrix**
- You want to **compare categories side-by-side**
- You have **duplicate combinations** and need aggregation
- You are preparing data for **analysis or modeling**

---

## 🔑 Key Insight

> Pivot transforms “rows of events” into a structured grid of values.

---

## 🧭 Mental Model

- Long format → messy, row-based
- Wide format → structured, column-based

`pivot_table` is the bridge between the two.

---

## ⚠️ Note

If no duplicates exist, you can use:

    data.pivot(...)

If duplicates exist, use:

    data.pivot_table(...)

---

## 💡 Final Takeaway

`pivot_table` makes your data:

- easier to read
- easier to compare
- ready for analysis

```python
data.pivot_table(index='date',columns='sector',values='signal1',aggfunc=np.mean)
```

```python
data.pivot_table(index='date',columns='sector',values='signal1',aggfunc=np.max)
```

# Pandas Merge: Conflicts, Edge Cases, and How to Handle Them

## 🧠 Core Idea

`merge()` combines datasets based on matching keys.

When conflicts occur, Pandas:
- **does not resolve them**
- **preserves all data**
- **forces you to decide what is correct**

---

## 🔧 Basic Example

    df = data.merge(data2, on=['sector','date'])

---

## ⚠️ Edge Case 1: Conflicting Column Values

### Problem

Same keys, different values:

| sector | date       | signal1 |
|--------|------------|--------|
| Tech   | 2021-01-01 | 0.5    |

| sector | date       | signal1 |
|--------|------------|--------|
| Tech   | 2021-01-01 | 0.9    |

---

### Result

| sector | date       | signal1_x | signal1_y |
|--------|------------|-----------|-----------|
| Tech   | 2021-01-01 | 0.5       | 0.9       |

---

### Why This Happens

- Pandas keeps both values to avoid silent data loss
- Adds suffixes:
  - `_x` → left DataFrame
  - `_y` → right DataFrame

---

### How to Handle

#### Choose one source:
    df['signal1'] = df['signal1_x']

#### Combine values:
    df['signal1'] = df[['signal1_x','signal1_y']].mean(axis=1)

#### Detect conflicts:
    (df['signal1_x'] != df['signal1_y']).sum()

---

## ⚠️ Edge Case 2: Duplicate Keys (One-to-Many)

### Problem

Left dataset has duplicate keys:

| sector | date       | signal1 |
|--------|------------|--------|
| Tech   | 2021-01-01 | 0.5    |
| Tech   | 2021-01-01 | 0.6    |

---

### Result

| sector | date       | signal1 | return |
|--------|------------|--------|--------|
| Tech   | 2021-01-01 | 0.5    | 0.01   |
| Tech   | 2021-01-01 | 0.6    | 0.01   |

---

### Why This Happens

- One row in right dataset matches multiple rows in left dataset

---

### How to Handle

#### Remove duplicates:
    data = data.drop_duplicates(subset=['sector','date'])

#### Aggregate first:
    data = data.groupby(['sector','date']).mean().reset_index()

---

## ⚠️ Edge Case 3: Many-to-Many Merge (Data Explosion)

### Problem

Duplicates exist in both datasets.

Example:
- 2 rows in left
- 3 rows in right

👉 Result: 6 rows

---

### Why This Happens

- Every matching row combines with every other matching row

---

### How to Handle

#### Validate merge relationships:
    data.merge(data2, on=['sector','date'], validate='one_to_one')

Other options:
    validate='one_to_many'
    validate='many_to_one'

---

## ⚠️ Edge Case 4: Missing Matches

### Problem

Some keys exist in one dataset but not the other.

---

### Behavior

Default (`how='inner'`):
- Drops unmatched rows

---

### How to Handle

#### Keep all left data:
    data.merge(data2, on=['sector','date'], how='left')

#### Keep everything:
    data.merge(data2, on=['sector','date'], how='outer')

---

## 🧠 Best Practices

### 1. Always check duplicates before merging

    data.duplicated(['sector','date']).sum()

---

### 2. Use `validate` to enforce expectations

    data.merge(data2, on=['sector','date'], validate='one_to_one')

---

### 3. Be explicit about columns

    data.merge(data2[['sector','date','return']], on=['sector','date'])

---

### 4. Inspect results after merge

    df.head()
    df.shape

---

## 🧭 Mental Model

`merge()` does not clean or fix data.

It:

> **combines datasets and exposes inconsistencies**

---

## 🔑 Final Takeaway

- Conflicts are expected
- Pandas preserves all data
- You must define how to resolve inconsistencies

---

## 💡 One Rule to Remember

> Always assume your data is messy and verify before and after every merge
