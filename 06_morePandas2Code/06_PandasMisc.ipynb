# Rolling Windows in Financial Analysis

## Overview

A **rolling window** (or moving window) is a technique used to compute statistics over a fixed subset of data that moves forward through time. Instead of calculating a single statistic over an entire dataset, rolling operations generate a **time series of statistics**, each based on the most recent observations.

This approach is fundamental in financial analysis, where market behavior changes over time and static metrics are often misleading.

---

## Conceptual Explanation

Given a time series:

```
t1, t2, t3, ..., tn
```

A rolling window of size `k` computes a statistic using:

```
[t1 → tk], then [t2 → t(k+1)], then [t3 → t(k+2)], ...
```

Each step shifts the window forward by one period.

---

## Example Dataset

```python
import pandas as pd
import numpy as np

univ = ['AAPL','MSFT','BAC','GS']
dates = pd.date_range('20110101','20201231')
df = pd.DataFrame(np.random.randn(len(dates),len(univ)), index=dates, columns=univ)
```

---

## Rolling Mean

```python
df.rolling(252).mean()
```

* Computes the average over a **252-day window** (~1 trading year)
* Returns `NaN` for the first 251 rows (insufficient data)

---

## Handling Early Data with `min_periods`

```python
df.rolling(252, min_periods=1).mean()
```

* Allows computation with fewer observations
* Early values are based on available data:

  * Day 1 → mean of 1 value
  * Day 10 → mean of 10 values
  * Day 252 → full rolling window

---

## Why Rolling Windows Matter in Finance

Financial data is **non-stationary**, meaning statistical properties change over time. Rolling windows allow analysis to adapt to recent conditions rather than relying on outdated historical averages.

---

## Common Financial Applications

### 1. Rolling Mean (Trend Analysis)

```python
df['AAPL'].rolling(252).mean()
```

* Smooths noise
* Used in moving averages
* Supports trend-following strategies (e.g., 50-day vs 200-day crossover)

---

### 2. Rolling Volatility (Risk Measurement)

```python
df['AAPL'].rolling(252).std()
```

* Measures variability of returns
* Used in:

  * Risk models
  * Portfolio allocation
  * Options pricing inputs

---

### 3. Rolling Returns

```python
df['AAPL'].pct_change().rolling(252).mean()
```

* Computes average return over a moving window
* Helps evaluate performance trends

---

### 4. Rolling Correlation (Diversification Analysis)

```python
df['AAPL'].rolling(252).corr(df['MSFT'])
```

* Tracks how relationships between assets evolve
* Critical because correlations are time-dependent

---

### 5. Rolling Sharpe Ratio (Risk-Adjusted Performance)

```python
returns = df['AAPL'].pct_change()
rolling_sharpe = returns.rolling(252).mean() / returns.rolling(252).std()
```

* Measures return per unit of risk
* Evaluates strategy quality over time

---

## Key Insight

Rolling statistics provide **local context**:

* Static metrics → long-term averages
* Rolling metrics → recent behavior

This distinction is critical in financial modeling, where recent data often carries more relevance than distant history.

---

## Important Considerations

### 1. Backward-Looking Nature

Rolling windows only use past data:

```
[t - window + 1, ..., t]
```

This prevents lookahead bias when used correctly.

---

### 2. Window Size Selection

* Small window → responsive but noisy
* Large window → stable but slow to adapt

Choosing the window size is a trade-off between sensitivity and stability.

---

### 3. Missing Data Behavior

* Default: requires full window → produces `NaN`
* With `min_periods`: allows partial calculations

---

## Summary Table of Rolling Operations

| Command                                | Description                                | Financial Use Case                      |
| -------------------------------------- | ------------------------------------------ | --------------------------------------- |
| `df.rolling(window)`                   | Creates a rolling window object            | Foundation for all rolling calculations |
| `.mean()`                              | Rolling average                            | Trend detection, moving averages        |
| `.std()`                               | Rolling standard deviation                 | Volatility / risk measurement           |
| `.var()`                               | Rolling variance                           | Alternative risk metric                 |
| `.sum()`                               | Rolling total                              | Volume aggregation, cumulative metrics  |
| `.min()` / `.max()`                    | Rolling extrema                            | Support/resistance levels               |
| `.corr(other)`                         | Rolling correlation                        | Asset relationship tracking             |
| `.cov(other)`                          | Rolling covariance                         | Portfolio risk modeling                 |
| `.apply(func)`                         | Custom rolling function                    | Advanced feature engineering            |
| `min_periods=`                         | Minimum observations required              | Controls early-period behavior          |
| `.pct_change()`                        | Period returns (often used before rolling) | Return-based analytics                  |
| `.rolling().mean() / .rolling().std()` | Rolling Sharpe Ratio                       | Risk-adjusted performance               |

---

## Final Note

Rolling windows transform static datasets into **dynamic analytical systems**. They are essential for any serious financial modeling, risk analysis, or quantitative strategy development.
