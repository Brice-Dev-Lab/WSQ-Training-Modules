## Understanding `.loc` with a MultiIndex DataFrame

This DataFrame was created using `pd.concat(data)`, which results in a **MultiIndex**:

- **Level 0 (outer index):** data type (e.g., `'earnings'`, `'ret_prev'`, `'pe'`, `'analyst_est'`)
- **Level 1 (inner index):** dates (e.g., `'20210107'`)

Each row is identified by a tuple:
`('ret_prev', '20210107')`

---

### Command Breakdown (following cell)

`multi.loc[('ret_prev','20210107'), ['SPY','TLT']]`

#### 1. `.loc`
- Label-based indexing (uses index names, not positions)

#### 2. Row Selection
`('ret_prev','20210107')`

- Selects the row where:
  - Level 0 = `'ret_prev'`
  - Level 1 = `'20210107'`

#### 3. Column Selection
`['SPY','TLT']`

- Selects only these two columns from that row

---

### Result

Returns the values for SPY and TLT for `'ret_prev'` on `'20210107'`:

```
SPY    1.262910
TLT    0.941813
```

---

### Key Takeaways

- `.loc` uses **labels**, not positions
- MultiIndex rows require a **tuple**: `(level_0, level_1)`
- Columns are selected like a normal list

````python
# multi.loc[('ret_prev','20210107'), ['SPY','TLT']]
#
# Explanation:
# This DataFrame uses a MultiIndex created by pd.concat(data), where:
#   - Level 0 index = data type (e.g., 'earnings', 'ret_prev', etc.)
#   - Level 1 index = date (e.g., '20210107')
#
# ('ret_prev','20210107') is a tuple that selects a specific row:
#   - 'ret_prev' → first index level
#   - '20210107' → second index level
#
# ['SPY','TLT'] selects only those columns from that row
#
# Result:
# Returns the SPY and TLT values for 'ret_prev' on '20210107'
#
# Key idea:
#   - .loc = label-based indexing
#   - MultiIndex requires tuple-based row selection

multi.loc[('ret_prev','20210107'),['SPY','TLT']]
````

### Selecting a Single Column from a MultiIndex DataFrame

```python
ser = multi['SPY']
```

---

### What This Does

- `multi` is a **DataFrame** with multiple columns (SPY, TLT, VXX, QQQ)
- Selecting a single column (`'SPY'`) returns a **Series**

#### Why a Series?
- A **DataFrame** is 2-dimensional (rows + columns)
- A **Series** is 1-dimensional (a single column of data)

So:
```python
multi['SPY']
```
returns a **Series** because only one column is selected.

---

### What the Result Looks Like

The resulting `ser` still keeps the **MultiIndex**:

```
earnings      20210105    0.8747
              20210106    0.5045
...
ret_prev      20210107    1.2629
...
pe            20210108    1.1127
...
analyst_est   20210109    2.6261
```

- Index = `(category, date)`
- Values = SPY data

---

### Purpose of Creating a Series

#### 1. Focus on a Single Variable
Instead of working with all columns:
```python
multi
```

You isolate one:
```python
ser
```

---

#### 2. Easier MultiIndex Slicing

```python
ser.loc['ret_prev']
```

Returns:
```
20210105   -0.6496
20210106    0.6976
20210107    1.2629
...
```

---

#### 3. Useful for Analysis and Modeling

Working with a Series makes it easier to:
- Compute statistics (`mean`, `std`)
- Apply transformations (`pct_change`, rolling averages)
- Use as input for models

Example:
```python
ser.mean()
ser.std()
ser.pct_change()
```

---

#### 4. Cleaner Data Manipulation

A Series simplifies operations compared to handling multiple columns at once.

---

### Important Detail

```python
multi['SPY']     # returns a Series
multi[['SPY']]   # returns a DataFrame
```

- Single brackets → Series
- Double brackets → DataFrame

---

### Key Takeaways

- Selecting one column returns a **Series**
- The **MultiIndex is preserved**
- Series are easier to work with for analysis and modeling
